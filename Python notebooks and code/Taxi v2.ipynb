```python
def decode(i):
    
    output = []
    output.append(i%4)
    i=round(i/4)
    output.append(i%5)
    i=round(i/5)
    output.append(i%5)
    i=round(i/5)
    output.append(i)
    assert 0<= i <5
    return output
```

```python
def random_policy(env):
    state = env.reset()
    total_reward = 0
    while True:
        action = env.action_space.sample()
        state, reward, done,info = env.step(action)
        if done:
            total_reward += reward
            print(total_reward)
            break
        else:
            env.render()
            sys.stdout.flush()
            total_reward += reward
            continue
```

```python
random_policy(env)
```

```python
def select_action(env,episode_i,q_s):
    
    epsilon = 1/episode_i
    
    probs = np.ones(nA)*(epsilon/nA)
    probs[np.argmax(q_s)] = (1-epsilon)+(epsilon/nA)
    
    action = np.random.choice(np.arange(nA),p=probs)
    
    return action
```

```python
def sarsa(env,num_episodes,alpha,gamma=1):
    
    q = defaultdict(lambda:np.zeros(nA))

    for episode_i in range(1,num_episodes+1):
        
        if episode_i%10==0:
            print("\rEpisode:{}/{}".format(episode_i,num_episodes),end="")
            sys.stdout.flush()
        state = env.reset()
        action = select_action(env,episode_i,q[state])
        for t_step in range(300):
            new_state,reward,done,info = env.step(action)
            if done:
                q[state][action] = q[state][action] + alpha*(reward - q[state][action])
                break
            
            new_action = select_action(env,episode_i,q[new_state])
            d_reward = reward + gamma*(q[new_state][new_action])
            q[state][action] = q[state][action] + alpha*(d_reward - q[state][action])
            
            state = new_state
            action = new_action
    return q
```

```python
sarsa(env,20000,0.02)
```

```python
def sarsa_max(env,num_episodes,alpha,gamma=1):
    
    q = defaultdict(lambda:np.zeros(env.action_space.n))
    
    for episode_i in range(1,num_episodes+1):
        
        if episode_i%100==0:
            print("\rEpsiode:{}/{}".format(episode_i,num_episodes),end="")
            sys.stdout.flush()
        
        #Initial state
        state = env.reset()
        
        #Initial action for the state
        action = select_action(env,episode_i,q[state])
        
        for t_step in range(300):
            new_state,reward,done,info = env.step(action)
            
            if done:
                q[state][action] = q[state][action] + alpha*(reward-q[state][action])
                break
                
            d_reward = reward + gamma*(np.amax(q[new_state]))
            q[state][action] = q[state][action] + alpha*(d_reward - q[state][action])
            state = new_state
            action = select_action(env,episode_i,q[state])
    return q
```

```python
sarsa_max(env,5000,0.02)
```
