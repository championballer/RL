```python
def decode(i):
    
    output = []
    output.append(i%4)
    i=round(i/4)
    output.append(i%5)
    i=round(i/5)
    output.append(i%5)
    i=round(i/5)
    output.append(i)
    assert 0<= i <5
    return output
```

```python
def random_policy(env):
    state = env.reset()
    total_reward = 0
    while True:
        action = env.action_space.sample()
        state, reward, done,info = env.step(action)
        if done:
            total_reward += reward
            print(total_reward)
            break
        else:
            env.render()
            sys.stdout.flush()
            total_reward += reward
            continue
```

```python
random_policy(env)
```
